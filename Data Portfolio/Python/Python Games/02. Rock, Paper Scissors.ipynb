# Intro

Rock, Paper, Scissors is a simple game. I don't think I need to explain it, but I will do so nonetheless. Two players (you can sometimes play in a group as well) face off picking either a rock, a paper, or a pair of scissors and making hand signs that represent their choice. If both pick the same, the game starts again. The winner is determined by the following:
* Rock beats Scissors
* Scissors beats Paper
* Paper beats Rock

In this case, we will have a player face off a computer that will randomly pick one choice. The game will have no graphics, but only text. 

# Importing Libraries 

The frist thing to do is help ourselves to using *shuffle* and *clear_output* since I want to be able to repeat the game if I wanted to.

```python
from random import shuffle
from IPython.display import clear_output
```

# Gameplay Mechanics

The mechanics will involve the human player choosing one of the options by inputing 1, 2 or 3. In case he picks another value, the code will ask again for a corect number. 

First things first, we have to create a dictionary which will enable us to link our choice with a value.
1. Rock
2. Paper
3. Scissors

```python
#Rock, Paper, Scissors starts with a dictionary. It's called hands because, usually when you play the game, you make a hand sign.

hands = {"Rock":"1", "Paper":"2", "Scissors":"3"}
```

## Player mechanics

Next, we have to create the structure on which a player will get to select a choice. As mentioned, if the choice is invalid, the player will be asked again for a choice.

```python
#Player choice mechanics

def player_move():
    
    """
    First the function will involved a choice, which is default set to 0. Then it will print out the options at hand.
    A loop will repeat until the player has an input between 1, 2 or 3.
    At the end, the value of your choice is displayed.
    """
    #Default choice
    choice = 0
    
    #Printing out options
    print('Choose:')
    print('Rock - 1')
    print('Paper - 2')
    print('Scissors - 3')
    
    #While input is not 1,2 or 3, the loop will repeat. 
    while choice not in ["1","2","3"]:
        #Value will be a string
        choice = input('Pick a number "1" or "2" or "3": \n')
    
    #Determining the correspondent to your choice and displaying it
    for p_key,p_value in hands.items():
        if choice == p_value:
            print(f'You have chosen: {p_key}')
            return p_key
```

## Computer mechanics

Next, we have to get the computer to randomly select something.

```python
# Computer choice mechanics

def computer_move():
    """
    In creating the computer move, I though that what this function will do is actually cast the dictionary keys as a list and shuffle it.
    The first item in the list will be its pick.
    """  
    
    # Casting the dictionary keys as a list
    computer_hands = list(hands.keys())
    
    #Shuffling the list
    shuffle(computer_hands)
    
    #Returning the first item of the list
    return computer_hands[0]
```

# Determining the Winner

As explained earlier, *Rock beats Scissors*, *Scissors beats Paper*, *Paper beats Rock*. We have to translate this into code. Of course, if both pick the same, it will be a draw.

```python
#Winner mechanics function

def win(player_choice, computer_choice):
    
    #This will basically be a big "if"
    
    #Draw
    if player_choice == computer_choice:
        print("It's a draw")
    
    #Rock 
    elif player_choice == "Rock" and computer_choice == "Paper":
        print("Computer Wins!")
    elif player_choice == "Rock" and computer_choice == "Scissors":
        print("Player Wins!")
    
    #Paper
    elif player_choice == "Paper" and computer_choice == "Rock":
        print("Player Wins!")
    elif player_choice == "Paper" and computer_choice == "Scissors":
        print("Computer Wins!")
    
    #Schissors
    elif player_choice == "Scissors" and computer_choice == "Rock":
        print("Computer Wins!")
    elif player_choice == "Scissors" and computer_choice == "Paper":
        print("Player Wins!")
```

## Repeat

If the player wants to go again, he/she should have that option, without having to restart the whole thing again.

```python
#Repeat function

def repeat(): 
    
    """
    This will be quite straightforward.
    We will have a list with two choices, "yes" and "no". You should be able to write it in caps lock, capitalized, lower case, either way. 
    As long if it's not "yes" or "no", the code will repeat.
    When "yes" is the input, the game will restart.
    When "no" is the input, the game will stop.
    
    """  
    
    #Creating the list and the choice
    repeat = ['yes','no']
    select = " "
    
    #Leaving two spaces just because we want to make it a little more visible for the user
    print("\n\n")
    
    
    #While select is not yes, or no, the code will repeat.
    while select not in repeat:
        select = input('Do you want to play again?\n Write "YES" or "NO":\n')
        select = select.lower() #transforming the choice into lower case
        if select not in repeat: #repeating the code in case of an invalid choice
            clear_output()
            print('Chose "YES" or "NO":\n')
    
    #If either yes or no are written, the function will determine if the game will return True or False
    if select == "yes":
        return True
    else:
        return False
```

## Game Intro

Lets make a game intro, so that we are able to explain the rules of the game.

```python
#Game Intro function will involve clearing the screen of any other inputs and outputs and presenting the rules of the game.

def game_intro():
    clear_output()
    print("Welcome to Rock Paper Scissors!")
    print('You know the rules:')
    print('Rock beats Paper')
    print('Paper beats Scissors')
    print('Scrissors beats Rock')
    print('Very fun...')
```

# THE GAME

The actual gameplay run will revolve around the cell below, which will ensure functions interact one with another.

```python
#game_on boolean will keep the game moving while it's true and will interact with the repeat() function
game_on = True

# Recreating the dictionary
hands = {"Rock":"1", "Paper":"2", "Scissors":"3"}

#Loop that will play the game as long as game_on is True

while game_on:
    
    #Game INTRO
    game_intro()
    
    #Waiting for the player to make the first move
    print('\nMake your move!')
    
    #Player choice and Computer Choice
    player_choice = player_move()
    computer_choice = computer_move()
    
    #Determining the winner
    win(player_choice, computer_choice)
    
    #Repeat
    game_on = repeat()

#End message for the game.
print('Thank you for playing! :)')
```
