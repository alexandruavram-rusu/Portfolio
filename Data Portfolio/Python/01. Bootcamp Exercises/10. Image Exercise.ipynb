```python
from PIL import Image
```

```python
words = Image.open("word_matrix.png")
words
```

```python
mask = Image.open("mask.png")
mask
```

First, we should resize the mask image to get it to fit the words image.

```python
# Getting words size

words.size
```

```python
# Resizing mask image

mask = mask.resize((1015,559))
mask
```

Next, we should make the mask image more transparent and then put it on top of the words image.

```python
# Multiple attempts

mask.putalpha(100)
mask
```

```python
mask.paste(words,box=(0,0),mask=mask)
```
