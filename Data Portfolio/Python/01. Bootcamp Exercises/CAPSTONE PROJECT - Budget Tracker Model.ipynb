# Database Design

Our budget tracker should have the following components:

1. Transaction category - either income or types of expenses, will be detailed below
2. Description - description of transaction
3. Transaction date - date of transaction
4. Amount - amount of transaction in local currency

## Transaction Categories

These are mostly designed for ease of use, so you might not have a full visual on your expenses.

01. Income - different sources of income
02. Household Goods - including groceries, takeout, drinks or other basic consumer goods (from soap to cleaning products)
03. Utilities - electricity, gas, water, internet
04. Loans - credit card or mortgage
05. Investments - savings accounts, investments in financial instruments
06. Entertainment - either subscriptions, games or going out
07. Transport - public or private
08. Travel - vacation and expenses incurred while on holiday
09. Medical - various medical expenses
10. Personal Care - such as going to get a haircut or getting your nails done
11. Personal Items - various stuff bought such as clothes or phones
12. Gifts - gifts & donations
13. Sports - expenses related to sporting activites or sportsware
14. Other expenses - other types of expenses


```python
 # Transaction categories dictionary
    
categories = {1:"Income",
              2:"Household Goods",
              3: "Utilities",
              4: "Loans",
              5: "Investments",
              6: "Entertainment",
              7: "Transport",
              8: "Travel",
              9: "Medical",
              10: "Personal Care",
              11: "Personal Items",
              12: "Gifts",
              13: "Sports",
              14: "Other Expenses"}
```

# Design the database

The database will be a class-type object which will hold various methods.

```python
class Database():
    
    # INIT METHOD
    
    def __init__(self):
        self.database = {}

    """
    Methods that Database class will hold:
    1. Transaction_today: transactions that are processed on current date
    2. Transaction: transactions that are processed on a custom date
    3. Delete Transaction: ability to delete transaction
    4. View overview
    5. Breakdown on income
    6. Breakdown on expenses
    """    

    
    #When transactions are done on the present day
    def add_transaction_today(self):
        
        #number of transcations, aka transaction ID
        if len(self.database) == 0:
            no = 1
        else:
            no = list(self.database.keys())[-1] + 1
        
       
    #Getting transaction type
        t_cat = 0

        clear_output()
        print("Categories:\n01.Income\n02.Household Goods\n03.Utilities\n04.Loans\n05.Investments\n06.Entertainment\n07.Transport\n08.Travel\n09.Medical\n10.Personal Care\n11.Personal Items\n12.Gifts\n13.Sports\n14.Other expenses")

        while t_cat not in list(range(1,15)):
            try:
                t_cat = int(input("Choose a transaction category: "))
            except:
                print("Invalid selection.")
                continue
            else:
                continue
    
        t_cat = categories[t_cat]
        
        #t_date is today
        t_date = date.today()
        
        #Description
        clear_output()
        print("Category:", t_cat)
        print("Date:", t_date)
        t_desc = input("Enter a description: ")
        
        #Amount
        clear_output()
        print("Category:", t_cat)
        print("Date:", t_date)
        print("Description:", t_desc)
        while True:
            try:
                t_amount = float(input("Amount: "))
            except:
                print("An error has occured! Please try again!")
                continue
            else:
                break
        
        self.database.update({no:[t_cat,t_date,t_desc,t_amount]})
    
    
    
    
    #When transactions are done on another day (in the past or future)
    
    def add_transaction(self):
        
        #number of transcations, aka transaction ID
        if len(self.database) == 0:
            no = 1
        else:
            no = list(self.database.keys())[-1] + 1
        
       
    #Getting transaction type
        t_cat = 0

        clear_output()
        print("Categories:\n01.Income\n02.Household Goods\n03.Utilities\n04.Loans\n05.Investments\n06.Entertainment\n07.Transport\n08.Travel\n09.Medical\n10.Personal Care\n11.Personal Items\n12.Gifts\n13.Sports\n14.Other expenses")

        while t_cat not in list(range(1,15)):
            try:
                t_cat = int(input("Choose a transaction category: "))
            except:
                print("Invalid selection.")
                continue
            else:
                continue
    
        t_cat = categories[t_cat]
        
        #Getting the date of the transaction
        clear_output()
        while True:
            try:
                t_year = int(input("Year: "))
                t_month = int(input("Month: "))
                t_day = int(input("Day: "))
                t_date = date(t_year,t_month,t_day)
            except:
                print("Invalid date!")
                continue
            else:
                break
        
        #Description
        clear_output()
        print("Category:", t_cat)
        print("Date:", t_date)
        t_desc = input("Enter a description: ")
        
        #Amount
        clear_output()
        print("Category:", t_cat)
        print("Date:", t_date)
        print("Description:", t_desc)
        while True:
            try:
                t_amount = float(input("Amount: "))
            except:
                print("An error has occured! Please try again!")
                continue
            else:
                break
        
        self.database.update({no:[t_cat,t_date,t_desc,t_amount]})
        
    def delete_transaction(self):
        #First, print out dataframe
        print(pd.DataFrame.from_dict(self.database,orient="index",columns=["Transaction Category","Date","Description","Amount (in L.C.)"]))

        # t_del is the transaction we are going to delete
        t_del = 0
        
        #Choosing the transaction
        while t_del not in list(self.database.keys()):
            try:
                t_del = int(input("Choose a transaction you wish to delete: "))
            except:
                print("Invalid selection.")
                continue
            else:
                continue
        
        #Making sure the user wants the transaction
        clear_output()
        print("Are you sure you want to delete transaction {} - {} : {} ".format(t_del,self.database[t_del][2],self.database[t_del][3]))
        print("Yes or No")
        answer = ""
        
        while answer.lower() not in ["yes","no"]:
            try:
                answer = input("")
            except:
                print("Invalid selection! Please try again!")
                continue
            else:
                break
        
        if answer.lower()=="yes":
            print("Transaction has been deleted!")
            self.database.pop(t_del)
        else:
            print("Transaction has not been deleted!")
            pass
    
    
    # Displaying methods
        
    def display_db(self):
        
        #First, we are going to create a dataframe
        #Second, create dataframes for income and expenses with total values
        #Third, create print layout
        
        df = pd.DataFrame.from_dict(self.database,orient="index",columns=["Transaction Category","Date","Description","Amount (in L.C.)"])
        df = df.sort_values(by="Date")
        
        df_income = df.loc[df['Transaction Category'] == "Income"]
        income = df_income["Amount (in L.C.)"].sum()
        
        df_expenses = df.loc[df['Transaction Category'] != "Income"]
        expenses = df_expenses["Amount (in L.C.)"].sum()
        
        
        print("Total Income: {}".format(income))
        print("Total Expenses: {}".format(expenses))
        print("Budget Balance: {}".format(income-expenses))
        print("\n")
        print(df)
        print("\n")
        
        #Graphs to show income/Expense evolution
        
        plt.figure(num = 3, figsize=(10,5))
        plt.plot(df_income["Date"], df_income["Amount (in L.C.)"])
        plt.title("Income")
        plt.show()
        
        print("\n")
        
        plt.figure(num = 3, figsize=(10,5))
        plt.plot(df_expenses["Date"], df_expenses["Amount (in L.C.)"])
        plt.title("Expenses")
        plt.show()
        
    # Income
    def db_income(self):
        df = pd.DataFrame.from_dict(self.database,orient="index",columns=["Transaction Category","Date","Description","Amount (in L.C.)"])
        df = df.sort_values(by="Date")
        
        df_income = df.loc[df['Transaction Category'] == "Income"]
        income = df_income["Amount (in L.C.)"].sum()
        
        print("Total Income: {}".format(income))
        print(df_income)
    
    # Expense breakdowns
    def db_expenses(self):
        df = pd.DataFrame.from_dict(self.database,orient="index",columns=["Transaction Category","Date","Description","Amount (in L.C.)"])
        df = df.sort_values(by="Date")
        
        df_expenses = df.loc[df['Transaction Category'] != "Income"]
        expenses = df_expenses["Amount (in L.C.)"].sum()
        
        print("Total Expenses: {}".format(expenses))
        print(df_expenses)
```

# Running the program

```python
# db is the Databse class type object

db = Database()
```

```python
# function to display choice

def intro():
    print("Welcome to Budget tracker. Where you can track your budget.")
    print("========================================")
    print("1. Add transaction made today")
    print("2. Add transaction made on another date")
    print("3. Delete transaction")
    print("4. Track overview")
    print("5. See income evolution")
    print("6. See expenses evolution")
    print("7. Exit")
```

```python
# Running the program

while True:
    clear_output()

    
    intro()
    print("\n")
    
    # Making our choice
    choice = "" 
    while choice not in list(range(1,8)):
        try:
            choice = int(input("Option: "))
        except:
            print("Invalid Selection! Please try again.")
            continue
        else:
            break
    
    # Linking our choice to the class methods
    clear_output()
    if choice == 1:
        db.add_transaction_today()
        print("\n")
    elif choice == 2:
        db.add_transaction()
        print("\n")
    elif choice == 3:
        db.delete_transaction()
        print("\n")
    elif choice == 4:
        db.display_db()
        print("\n")
    elif choice == 5:
        db.db_income()
        print("\n")
    elif choice == 6:
        db.db_expenses()
        print("\n")
    elif choice == 7:
        break
    
    input("Press ENTER to continue.")
    
    
    
print("Thank you for using the budget tracker!")
```
